```python
import camelot
import pandas as pd
import glob, os
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
path = r"C:\Users\user\Desktop\ADNOC\NLP\1283\2021-well-16a78-32-daily-reports\2021-well-16a78-32-daily-reports\drilling"
def readfiles(path):
    os.chdir(path)
    pdfs = []
    for file in glob.glob("*.pdf"):
#         print(file)
        pdfs.append(file)
    return pdfs
pdfs = readfiles(path)
```

```python
def readfile(file_name):
    tables = camelot.read_pdf(file_name)
    df = tables[0].df
    return df
```

```python
def generate_report_details(df, report_name):
    df2 = df.iloc[0:5 ,:].drop(1)
    
    test = ""
    for index,row in df2.iterrows():
        test=test+'\n'+str(row[0])
    
    x = test.split("\n")
    x2 = [item for item in x if ":" in item]
    x3 = [item for item in x2 if not item.endswith(":")]

    my_dict = {}
    
    for item in x3:
        key, value = item.split(':', 1)
        my_dict[key] = value
    
    my_dict['Report_title'] = report_name
    return my_dict
```

```python
def generate_report_event_details(df):
    df = df.drop_duplicates()
    index1 = df[df[0] == 'FROM\nTO\nHRS\nPHASE\nCODE\nNPT\nOPERATIONS'].index.item()
    index2 = df[df[0] == 'TOTAL HRS'].index.item()
    sliced_df = df.loc[index1+1:index2-1]
#     sliced_df.replace('', np.nan, inplace=True)
#     sliced_df = sliced_df.dropna(axis=1, how='all')
    sliced_df = sliced_df.iloc[: , :10]
    print (len(sliced_df))
    try:
        sliced_df = sliced_df.drop([4,6,8], axis=1)
    except:
        pass
    print (sliced_df.shape)
    sliced_df.set_axis(['FROM','TO','HRS','PHASE','CODE','NPT','OPERATIONS'], axis='columns', inplace=True)
    sliced_df = sliced_df.reset_index()
    return sliced_df
```

```python
def join_report_details_events(report_det_dict, report_det_event):
    new_dict = pd.DataFrame(report_det_dict, index=[0])
    new_dict = pd.concat([new_dict]*report_event_details.shape[0], ignore_index=True)

    result = pd.concat([new_dict, report_det_event], axis=1)
    return result
```

```python
num_of_pdfs = range(len(pdfs))
for i in num_of_pdfs:
    print (f"{i} of {str(num_of_pdfs)}")
    report_name = pdfs[i]
    print (report_name)
    df = readfile(report_name)
    report_details = generate_report_details(df, report_name)
    report_event_details = generate_report_event_details(df)
    print (report_event_details.shape)
    result = join_report_details_events(report_details,report_event_details)
    report_name = report_name[:-4]
    result.to_csv(f"{report_name}.csv", index = False)
#     result.to_csv(fr"C:\Users\user\Desktop\ADNOC\NLP\1283\generated_reports\"+{report_name[2:-4]}.csv", index = False)
#     break
```
